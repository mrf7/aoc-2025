```kotlin
import com.toldoven.aoc.notebook.AocClient
%use kandy
%use dataframe

val aoc = AocClient.fromEnv().interactiveDay(2025, 2)
aoc.viewPartOne()

val exampleInput =
    "11-22,95-115,998-1012,1188511880-1188511890,222220-222224,1698522-1698528,446443-446449,38593856-38593862,565653-565659,824824821-824824827,2121212118-2121212124"
```

```kotlin
fun solvePartOne(input: String): Long {
    return input.split(",").map { it.split("-") }
        .map { (first, second) -> first.toLong()..second.toLong() }
        .also { println(it) }
        .flatMap { range ->
            range.filter {
                val id = it.toString()
                id.take(id.length / 2) == id.drop(id.length / 2)
            }
        }.sum()
}
solvePartOne(exampleInput)
```

```kotlin
solvePartOne(aoc.input())
aoc.submitPartOne(solvePartOne(aoc.input()))
```

```kotlin
fun isInvalid(id: Long): Boolean {
    val str = id.toString()
    return (1..str.length / 2).any { size ->
        (str.chunked(size).toSet().size == 1).also { if (it) DISPLAY(" $id: ${str.chunked(size)}") }
    }
}

fun solvePartTwo(input: String): Long {
    return input.split(",").map { it.split("-") }
        .map { (first, second) -> first.toLong()..second.toLong() }
//        .also { println(it) }
        .flatMap { range ->
            range.filter {
                isInvalid(it)
            }
        }.sum()
}
solvePartTwo(exampleInput)
```

```kotlin
aoc.submitPartTwo(solvePartTwo(aoc.input()))
```
