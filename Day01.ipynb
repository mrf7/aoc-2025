```kotlin
import io.kotest.assertions.assertSoftly
import io.kotest.matchers.shouldBe

val exampleAnswer = 3
val exampleInput = """L68
L30
R48
L5
R60
L55
L1
L99
R14
L82"""

fun Int.modAdd(other: Int, max: Int = 99, min: Int = 0) = (this + other).let {
//    (this + (other%100)).let { if (it < min) it + max + 1 else if (it > max) it - (max + 1) else it }
    (this + (other%100)).let { if (it !in min..max) it - (it.sign * (max+1)) else it }
}
assertSoftly {
    99.modAdd(1) shouldBe 0
    5.modAdd(-10) shouldBe 95
    0.modAdd(-1) shouldBe 99
    0.modAdd(-100) shouldBe 0
    0.modAdd(-603) shouldBe 97
}
"Pass"
```

```kotlin
import io.kotest.matchers.collections.shouldBeIn
import io.kotest.matchers.ints.shouldBeInRange

fun solvePartOne(input: List<String>, start: Int = 50): Int {
    return input
        .map { it.first() to it.drop(1).toInt() }
        .runningFold(start) { acc, (dir, dist) ->
            acc.modAdd(if (dir == 'R') dist else -dist).also { it.shouldBeInRange(0..99) }
        }
//        .also { DISPLAY(it) }
        .count { it == 0 }
}
solvePartOne(exampleInput.lines()) shouldBe exampleAnswer
```

```kotlin
import io.kotest.assertions.assertSoftly
import io.kotest.matchers.shouldBe

fun countSpins(current: Int, dist: Int): Pair<Int, Int> =
    (current + (dist % 100)).let {
        if (it !in 0..99) {
            it - (it.sign * 100) to 1 + (dist / 100)
        } else {
            it to 0 + (dist / 100)
        }
    }
countSpins(50, -68)
assertSoftly {
    countSpins(50, -68) shouldBe (82 to 1)
    countSpins(95, 60) shouldBe (55 to 1)
    countSpins(0, 1000) shouldBe (0 to 10)
}
```

```kotlin
val partTwoAnswer = 6
fun solvePartTwo(input: List<String>) =
    input
        .map { if (it.first() == 'R') it.drop(1).toInt() else -it.drop(1).toInt() }
        .runningFold(50 to 0) { (curr, count), dist ->
            val res = countSpins(curr, dist)
            res.first to count + res.second
        }
solvePartTwo(exampleInput.lines()).last().second shouldBe partTwoAnswer
solvePartTwo(exampleInput.lines())
```

```kotlin
val sol = solvePartTwo(aoc.input().lines())
```
