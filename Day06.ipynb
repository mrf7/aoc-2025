```kotlin
import com.toldoven.aoc.notebook.AocClient
%use kandy
%use dataframe

val aoc = AocClient.fromEnv().interactiveDay(2025, 6)
```

```kotlin
val exampleInput = """
123 328  51 64
 45 64  387 23
  6 98  215 314
*   +   *   +  """.trimIndent()
```

```kotlin
fun String.toOperation(): (Long, Long)-> Long = when (this) {
    "+" -> Long::plus
    "*" -> Long::times
    else -> error("invalid op $this")
}
```

```kotlin
fun solvePartOne(input: String): Long {
    val input = input.lines().map { it.split(' ').filter { it.isNotBlank() } }
    val (nums, ops) = input.dropLast(1) to input.last()
    return ops.mapIndexed { index, string ->
        val op: (Long, Long) -> Long = string.toOperation()
        nums.map { it[index].toLong() }.reduce(op)
    }.sum()
}
```

```kotlin
import io.kotest.matchers.shouldBe

solvePartOne(exampleInput) shouldBe 4277556
```

```kotlin
aoc.submitPartOne(solvePartOne(aoc.input()))
```

```kotlin
import io.kotest.mpp.start

fun solvePart2(input: String): Long {
    val input = input.lines()
    val (numLines, ops) = input.dropLast(1) to input.last()
    val opIndices = ops.mapIndexedNotNull { index, ch -> index.takeIf { ch in listOf('+', '*') }?.let { it to ch } }
    val probs: List<Pair<List<String>, Char>> = opIndices.windowed(2, partialWindows = true).map { operations ->
        val startIndex = operations.first().first
        val endIndex = operations.getOrNull(1)?.first
        val numbers = numLines.map { line ->
            if (endIndex != null) {
                line.slice(startIndex..endIndex - 2)
            } else {
                line.substring(startIndex)
            }.replace(' ', '_').reversed()
        }
        numbers to operations.first().second
    }
    val withNums: List<Pair<List<Long>, Char>> = probs.mapFirst { lines ->
        lines.first().indices.map { col ->
            lines.mapNotNull { it[col].takeIf { it.isDigit() } }.joinToString("").toLong()
        }
    }
    return withNums.sumOf { (nums, operation) ->
        nums.reduce(operation.toString().toOperation())
    }
}

solvePart2(exampleInput) shouldBe 3263827L
```

```kotlin
solvePart2(aoc.input())
```

```kotlin
aoc.submitPartTwo(solvePart2(aoc.input()))
```
